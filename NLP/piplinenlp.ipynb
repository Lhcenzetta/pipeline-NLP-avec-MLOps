Load the dataset

```python
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()
path_data = os.getenv("new_data")
df = pd.read_csv(path_data)
df
```

Normalisation (lowercae)

```python
df = df.map(lambda x: x.lower() if isinstance(x,str) else x)
df
```

remove pounctuation mark


```python
import string

translator = str.maketrans('', '', string.punctuation)

df = df.apply(lambda col: col.map(lambda x: str(x).translate(translator) if isinstance(x, str) else x)) 

df
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize

df["body_cols"] = df["body_cols"].apply(lambda x: word_tokenize(str(x)))

df
```

```python
nltk.download('stopwords')
```

```python
import nltk
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

df['body_cols'] = df['body_cols'].apply(lambda x: [word for word in x if word not in stop_words])
df
```
