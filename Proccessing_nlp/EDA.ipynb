Load the data

```python
import os
from dotenv import load_dotenv
import pandas as pd

load_dotenv()
path_data = os.getenv("data_path")
df = pd.read_csv(path_data)
df.info()
```

```python
df.isnull().sum()
```

```python
df["subject"] = df["subject"].fillna("")
df["body"] = df["body"].fillna("")
df["answer"] = df["answer"].fillna("")
```

```python
df.isnull().sum()
```

```python

tag_cols = ["tag_1", "tag_2", "tag_3", "tag_4", "tag_5", "tag_6", "tag_7", "tag_8"]
df["full_tag"] = df[tag_cols].apply(
    lambda x: ",".join(x.dropna().astype(str)),
    axis=1
)
df = df.drop(columns=tag_cols)

df
```

```python
body_cols = ["subject",	"body"]
df["body_cols"] = df[body_cols].apply(
    lambda x: ",".join(x.dropna().astype(str)),
    axis=1
)

df = df.drop(columns= body_cols)
df.info()
```
